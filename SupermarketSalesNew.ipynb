```python
import pandas as pd
df = pd.read_csv("supermarket_sales_new.csv")

#Before Cleaning
df.head()
df.info()
df.isna().sum()
df.duplicated().sum()
df["expected_tax"] = df["Unit price"] * df["Quantity"] * 0.05
df[["Tax 5%", "expected_tax"]].head()
df = df.drop(columns=["expected_tax"])
df["Total"] = df["Unit price"] * df["Quantity"] + df["Tax 5%"]
df.to_csv("supermarket_sales_cleaned.csv", index=False)

#After Cleaning
df_cleaned = pd.read_csv("supermarket_sales_cleaned.csv")
df_cleaned.head()

df_cleaned.describe()
df_cleaned['City'].value_counts()
df_cleaned['Product line'].value_counts()
df_cleaned[['Unit price','Quantity','Tax 5%']].corr()

### Data Quality Summary
#No missing values.
#No duplicated rows.
#Data types are appropriate.
#One column had unexpected tax which was recalculated.
#Total field was recalculated for accuracy.
```
