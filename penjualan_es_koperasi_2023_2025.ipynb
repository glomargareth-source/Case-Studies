```python
import pandas as pd

df = pd.read_csv("penjualan_es_2023_2025.csv")
df["Tanggal"] = pd.to_datetime(df["Tanggal"],errors="coerce")
df = df.dropna(subset=["Tanggal"])
text_cols = ["Jenis_pelanggan", "Musim", "Cuaca", "Hari_dalam_minggu"]
for col in text_cols :
    df[col] = df[col].astype(str).str.strip().str.lower()
df = df.drop_duplicates()
df = df.sort_values("Tanggal")

df["Total_es_terjual_kg"] = pd.to_numeric(df["Total_es_terjual_kg"], errors="coerce")
df["Jumlah_transaksi"] = pd.to_numeric(df["Jumlah_transaksi"], errors="coerce")

df.loc[df["Jumlah_transaksi"] < 0, "Jumlah_transaksi"] = None

df["Keterangan"] = df["Keterangan"].fillna("")

df.to_csv("penjualan_es_koperasi_2023_2025_clean.csv",index=False)
df.head()
```
